## My Lane as an ML Task

My chosen lane is **Search Intelligence – Content Refresh Prioritization**.

I would frame this primarily as a **ranking and scoring task**. The objective is to assign a priority score to each webpage and rank the pages from highest to lowest priority. This ranking helps the SEO and content team identify which pages should be reviewed first for possible content updates or optimization. The model is designed to support human decision-making rather than replace it.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

The target is to predict whether a webpage is showing signs of declining search performance. I use **is_declining_label** as the target, which is derived from the **trend_direction** column in the starter dataset. Pages with a **down** trend are labeled as **1 (declining)**, while all other pages are labeled as **0 (not declining)**.

This label is used as a practical proxy rather than a direct "needs refresh" indicator. A declining page may require further review, but the final decision should always be made by the SEO and content team.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

The primary success metric is **Precision@20**.

The SEO team can review only a limited number of pages at a time, so the quality of the top recommendations is more important than overall accuracy. A higher Precision@20 means that more of the highest-ranked pages are genuinely worth reviewing. This makes the model more useful for prioritizing content updates efficiently.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why Machine Learning Beats a Fixed Rule

A fixed rule can only evaluate a few manually selected conditions, such as content age or impressions. However, search performance is influenced by multiple interacting factors, including impressions, CTR, average position, content freshness, and update history.

Machine learning can identify patterns across these signals and estimate review priority more effectively than a single fixed rule. While a rule provides a useful baseline, an ML model can support more informed and data-driven decisions. The final decision should still be reviewed by the SEO and content team before any content changes are made.
